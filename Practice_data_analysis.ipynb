It is very hard to carry out operation on multiple columns without the NumPy library. Using stuff like lambda it can be tough to incorporate multiple conditions, and multiple columns into a data transformation. But NumPy makes this easy.

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df4 = pd.DataFrame(data)

condition1 = df4['A'] < 3
function1 = lambda x: x * 2
condition2 = df4['A'] >= 3
function2 = lambda x: x + 5

df4['C'] = np.where(condition1, function1(df4['B']), function2(df4['B']))

print(df4)
```

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df4 = pd.DataFrame(data)

condition1 = df4['A'] < 3
def function1(x,y):
    return x/(y*2)

condition2 = df4['A'] >= 3
function2 = lambda x: x + 5

df4['C'] = np.where(condition1, function1(df4['B'],df4['A']), function2(df4['B']))

print(df4)
```

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
# Define the conditions and corresponding functions/logic
condition1 = df['A'] < 3
function1 = lambda x: x * 2
condition2 = df['A'] >= 3
function2 = lambda x: x + 5
# Apply the functions/logic based on conditions and create the new column
df['C'] = np.where(condition1,function1(df['A']),function2(df['B']))
df
```

```python
#one way to iterate through a dataframe

 #first establish the first column to iterate through
#working_column = 0
    #Have for loop that goes through all the columns
def check_for_unusual_values(df):
    for column in range(0,(df.shape[1])):
    #Check if column is quantitative
        if (type(df.iloc[0][column]) == np.float64) or (type(df.iloc[0][column]) == np.int64):
    #THEN go into for loop going down through all rows in that column
            for row in range(0, (df.shape[0])):
                if df.iloc[row][column] > 9.9:
                #print("unusual value detected at "+str(row)+","+str(column))
                    print(f"unusual value detected at {row},{column}")
```

```python
def string_finder(df):
    for column in range(0,(df.shape[1])):
    #Check if column is NOT quantitative
        if (type(df.iloc[0][column]) != np.float64) or (type(df.iloc[0][column]) != np.int64):
    #THEN go into for loop going down through all rows in that column
            for row in range(0, (df.shape[0])):
                if type(df.iloc[row][column]) == str:
                #print("unusual value detected at "+str(row)+","+str(column))
                    print(f"string found at {row},{column}")
```
